```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\DELL\\OneDrive\\Desktop\\minorsem5\\training_set.csv')
```

```python
df.head()
```

```python
df['Service_Hour_Type']=df['Service hours type']
```

```python
df=df.drop('Service hours type', axis=1)
```

```python
df.head()
```

```python
if 'Service_Hour_Type' in df.columns:
    lemmatized_texts =[]

    for row in df['Service_Hour_Type']:
        if isinstance(row, str):
            doc=nlp(row)
            lemma_SHT=" ".join([token.lemma_ for token in doc])
        else:
            lemma_SHT=""
        lemmatized_texts .append(lemma_SHT)

    df['Service_Hour_Type']=lemma_SHT
    df.head()
else:
    print("Column 'Service_Hour_Type' not found in the DataFrame.")
```
